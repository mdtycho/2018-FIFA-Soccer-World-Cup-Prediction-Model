```python
def get_row(odds):
    probs = [1/(1+o) for o in odds]
    probs = [p/sum(probs) for p in probs]
    goalDiff = 0
    if probs[0] > probs[1]:
        if(probs[0]> probs[2]):
            goalDiff = 1
        else:
            goalDiff=0
    else:
        goalDiff = -1
    return [goalDiff] + probs
```
