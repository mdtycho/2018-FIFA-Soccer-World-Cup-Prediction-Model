```python
all_ranks = all_ranks[['date', 'home_team', 'away_team', 'home_score', 'away_score', 'neutral', 'year',
                        'home_rank', 'away_rank']]
```

```python
all_ranks.head()
```

```python
all_ranks['goal_diff'] = all_ranks['home_score'] - all_ranks['away_score']
```

```python
all_ranks['results'] = all_ranks['goal_diff'].apply(lambda x: (x/abs(x)) if x != 0 else 0)
```

```python
all_ranks.head()
```

```python
X = all_ranks[['neutral', 'home_rank', 'away_rank']]

y_clf = all_ranks['results']

y_reg = all_ranks['goal_diff']
```

```python
X['neutral'] = X['neutral'].astype(int)
```

```python
X.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train1, X_test1, y_train1, y_test1 = train_test_split(X, y_clf, test_size=0.30)

X_train2, X_test2, y_train2, y_test2 = train_test_split(X, y_reg, test_size=0.30)
```

```python
from sklearn.neural_network import MLPClassifier, MLPRegressor
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()

clf = MLPClassifier()

reg = MLPRegressor()
```

```python
# use pipeline in future

clf.fit(X_train1, y_train1)

reg.fit(X_train2, y_train2)
```

```python
p1 = clf.predict(X_test1)

p2 = reg.predict(X_test2)
```

```python
from sklearn.metrics import f1_score, accuracy_score, mean_squared_error
```

```python
print(f1_score(y_test1, p1, average = 'weighted'))
print(accuracy_score(y_test1, p1))

print(mean_squared_error(y_test2, p2))
```
